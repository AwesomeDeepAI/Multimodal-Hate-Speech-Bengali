```python
import pandas as pd
import re
from sklearn.utils import shuffle
from google.colab import drive
pd.set_option('display.max_colwidth', 255)
drive.mount('/content/drive')
```

```python
path = '/content/drive/My Drive/Multimodal-Hate-Speech-Bengali/'
train = pd.read_csv(f"{path}data/train.csv")
validate = pd.read_csv(f"{path}data/validate.csv") #Hateful_1324.jpg,"আর্মিতে পরীক্ষা দিতে আসছি, পাছা দেখাতে নয় । ",0,Neutral  i convert label 0 to 1 
test = pd.read_csv(f"{path}data/test.csv")
```

```python
len(train), len(validate), len(test), len(train) + len(validate) + len(test) 
```

```python
train = train.rename(columns={'Text': 'text', 'label': 'target',})
validate = validate.rename(columns={'Text': 'text', 'label': 'target',})
test = test.rename(columns={'Text': 'text', 'label': 'target',})


train['label'] = train['Images'].replace('_.*','',regex=True)
validate['label'] = validate['Images'].replace('_.*','',regex=True)
test['label'] = test['Images'].replace('_.*','',regex=True)

train = train[['text','label','target']]
validate = validate[['text','label','target']]
test = test[['text','label','target']]
```

```python
train.label.value_counts(), train.target.value_counts()
```

```python
test.label.value_counts(), test.target.value_counts()
```

```python
validate.label.value_counts(), validate.target.value_counts()
```

```python
for _ in range(5000):
  train = shuffle(train)
  validate = shuffle(validate)
  test = shuffle(test)
  train = train.reset_index(drop=True)
  validate = validate.reset_index(drop=True)
  test = test.reset_index(drop=True)
```

```python
train.to_csv(f"{path}data/train_text.csv")
validate.to_csv(f"{path}data/validate_text.csv")
test.to_csv(f"{path}data/test_text.csv")
```
